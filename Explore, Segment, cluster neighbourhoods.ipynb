```python
import pandas as pd
```

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

```python
df_red=df[df.Borough != 'Not assigned'] #ignore cells with borough not assigned
```

```python
df_grp=df_red[['Postcode','Borough','Neighbourhood']].groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
```
